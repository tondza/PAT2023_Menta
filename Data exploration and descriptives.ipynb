```python
# Set the style for seaborn
sns.set_style("whitegrid")

# Loop through each DASS score to create separate figures
for dass_score in dass_scores:
    # Create a figure with subplots
    fig, axes = plt.subplots(1, 5, figsize=(25, 5))  # Adjust the size as needed
    fig.suptitle(f'Heatmaps of TIPI Dimensions vs. {dass_score}', fontsize=16)

    # Loop through each TIPI dimension and create a heatmap
    for i, tipi_dimension in enumerate(tipi_dimensions):
        # Calculate the crosstab for the current TIPI dimension and DASS score
        joint_distribution = pd.crosstab(df_filtered[tipi_dimension], df_filtered[dass_score])

        # Plot the heatmap
        sns.heatmap(joint_distribution, annot=True, fmt='d', cmap='coolwarm', ax=axes[i])
        axes[i].set_title(tipi_dimension)
        axes[i].set_xlabel(dass_score)
        axes[i].set_ylabel('Score')
        
        # Invert the y-axis to have higher scores at the bottom
        axes[i].invert_yaxis()

    # Adjust the layout
    plt.tight_layout()
    # Adjust top margin to accommodate the suptitle
    plt.subplots_adjust(top=0.85)

    # Show the plot
    plt.show()
```

```python
tipi_dimensions = ['Extraversion', 'Agreeableness', 'Conscientiousness', 'Emotional_Stability', 'Openness']
dass_scores = ['Anxiety_Score', 'Stress_Score', 'Depression_Score']

# Combine the lists if you want to drop rows where either has NaNs
columns_to_check = tipi_dimensions + dass_scores

# Check if the columns exist in the DataFrame
missing_columns = [col for col in columns_to_check if col not in df_filtered.columns]
if missing_columns:
    print("These columns are missing in the DataFrame:", missing_columns)
else:
    # Drop rows where any of the specified columns have NaN
    corr_df = df_filtered.dropna(subset=columns_to_check)
```

```python
import pandas as pd
from scipy.stats import spearmanr

# Assuming df_filtered is your DataFrame
# Define the TIPI dimensions and DASS scores
tipi_dimensions = ['Extraversion', 'Agreeableness', 'Conscientiousness', 'Emotional_Stability', 'Openness']
dass_scores = ['Anxiety_Score', 'Stress_Score', 'Depression_Score']

# Create a DataFrame to store the Spearman's rho and p-values
correlation_matrix = pd.DataFrame(index=tipi_dimensions, columns=dass_scores)
p_value_matrix = pd.DataFrame(index=tipi_dimensions, columns=dass_scores)

# Calculate Spearman's rho and p-values for each combination
for tipi_dim in tipi_dimensions:
    for dass_score in dass_scores:
        rho, p_value = spearmanr(corr_df[tipi_dim], corr_df[dass_score])
        correlation_matrix.at[tipi_dim, dass_score] = rho
        p_value_matrix.at[tipi_dim, dass_score] = p_value

# Set display precision for floating point numbers
pd.set_option('display.float_format', '{:.6f}'.format)

# Display the correlation matrix
print("Spearman's Rho Correlation Matrix:")
print(correlation_matrix)

# Display the p-value matrix
print("\nP-value Matrix:")
print(p_value_matrix)
```

```python
import pandas as pd

# Assuming df_filtered is your DataFrame

# Function to calculate and print percentages for a given column
def print_percentage_distribution(column):
    percentages = df_filtered[column].value_counts(normalize=True).mul(100).round(2)
    print(f"Percentage Distribution for {column}:\n{percentages}\n")

# Calculate and print the distributions
print_percentage_distribution('gender_label')
print_percentage_distribution('race_label')
print_percentage_distribution('education_label')
print_percentage_distribution('country')
print_percentage_distribution('urban_label')
print_percentage_distribution('married_label')
print_percentage_distribution('religion_label')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Assuming df_filtered contains 'age_category' and 'gender_label' columns

# Order the age categories
age_order = ['<18', '18-35', '35-59', '60+']

# Calculate the count distribution of age categories by gender
age_gender_distribution = df_filtered.groupby(['age_category', 'gender_label']).size().unstack().fillna(0)
age_gender_distribution = age_gender_distribution.loc[age_order]  # Order the age categories

# Create a stacked bar plot
sns.set_style("whitegrid")
plt.figure(figsize=(7, 4))
age_gender_distribution.plot(kind='bar', stacked=True, color=['lightpink', 'skyblue', 'lightgrey'], ax=plt.gca())
plt.title('Distribution of Age Categories and Gender')
plt.xlabel('Age Category')
plt.ylabel('Count')
plt.xticks(rotation=0)  # Adjust rotation if necessary
plt.legend(title='Gender')
plt.show()
```
