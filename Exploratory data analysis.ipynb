# Calculating DASS questionaire scores for depression, anxiety and stress #

```python
# Calculate DASS scores

# Define the questions for each scale
depression_questions = ['Q3A', 'Q5A', 'Q10A', 'Q13A', 'Q16A', 'Q17A', 'Q21A', 'Q24A', 'Q26A', 'Q31A', 'Q34A', 'Q37A', 'Q38A', 'Q42A']
anxiety_questions = ['Q2A', 'Q4A', 'Q7A', 'Q9A', 'Q15A', 'Q19A', 'Q20A', 'Q23A', 'Q25A', 'Q28A', 'Q30A', 'Q36A', 'Q40A', 'Q41A']
stress_questions = ['Q1A', 'Q6A', 'Q8A', 'Q11A', 'Q12A', 'Q14A', 'Q18A', 'Q22A', 'Q27A', 'Q29A', 'Q32A', 'Q33A', 'Q35A', 'Q39A']

#Orignal scoring goes from 0 to 3. However, in the data it goes 1 to 4. Subtract one from every questions to calculate scoring correctly.
df[depression_questions] = df[depression_questions] - 1
df[anxiety_questions] = df[anxiety_questions] - 1
df[stress_questions] = df[stress_questions] - 1

# Calculate the scores for each scale
df['Depression_Score'] = df[depression_questions].sum(axis=1)
df['Anxiety_Score'] = df[anxiety_questions].sum(axis=1)
df['Stress_Score'] = df[stress_questions].sum(axis=1)

```

# Categorizing depression, anxiety and stress scores (based on DASS instructions trhesholds), visualizing distribution.

```python
# Estimate condition severity - defining categories and labels

def categorize_depression(score):
    if score <= 9:
        return 'Normal'
    elif score <= 13:
        return 'Mild'
    elif score <= 20:
        return 'Moderate'
    elif score <= 27:
        return 'Severe'
    else:
        return 'Extremely Severe'

def categorize_anxiety(score):
    if score <= 7:
        return 'Normal'
    elif score <= 9:
        return 'Mild'
    elif score <= 14:
        return 'Moderate'
    elif score <= 19:
        return 'Severe'
    else:
        return 'Extremely Severe'

def categorize_stress(score):
    if score <= 14:
        return 'Normal'
    elif score <= 18:
        return 'Mild'
    elif score <= 25:
        return 'Moderate'
    elif score <= 33:
        return 'Severe'
    else:
        return 'Extremely Severe'

# Apply the functions to each row in the df
df['Depression_Severity'] = df['Depression_Score'].apply(categorize_depression)
df['Anxiety_Severity'] = df['Anxiety_Score'].apply(categorize_anxiety)
df['Stress_Severity'] = df['Stress_Score'].apply(categorize_stress)

```

```python
# Calculate frequencies and percentages
depression_counts = df['Depression_Severity'].value_counts().sort_index()
anxiety_counts = df['Anxiety_Severity'].value_counts().sort_index()
stress_counts = df['Stress_Severity'].value_counts().sort_index()

depression_pct = df['Depression_Severity'].value_counts(normalize=True).sort_index() * 100
anxiety_pct = df['Anxiety_Severity'].value_counts(normalize=True).sort_index() * 100
stress_pct = df['Stress_Severity'].value_counts(normalize=True).sort_index() * 100

# Concatenate counts and percentages
depression_summary = pd.concat([depression_counts, depression_pct], axis=1, keys=['Counts', 'Percentage'])
anxiety_summary = pd.concat([anxiety_counts, anxiety_pct], axis=1, keys=['Counts', 'Percentage'])
stress_summary = pd.concat([stress_counts, stress_pct], axis=1, keys=['Counts', 'Percentage'])

# Display the concatenated dataframes
print("Depression Severity:")
print(depression_summary)
print("\nAnxiety Severity:")
print(anxiety_summary)
print("\nStress Severity:")
print(stress_summary)

# Descriptive statistics of the scores
depression_stats = df['Depression_Score'].describe()
anxiety_stats = df['Anxiety_Score'].describe()
stress_stats = df['Stress_Score'].describe()

# Print the descriptive statistics
print("\nDepression Score Statistics:\n", depression_stats)
print("\nAnxiety Score Statistics:\n", anxiety_stats)
print("\nStress Score Statistics:\n", stress_stats)

```

```python
# Pie chart for Depression Severity
depression_severity_counts = df['Depression_Severity'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(depression_severity_counts, labels=depression_severity_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Depression Severity Distribution')
plt.show()

# Pie chart for Anxiety Severity
anxiety_severity_counts = df['Anxiety_Severity'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(anxiety_severity_counts, labels=anxiety_severity_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Anxiety Severity Distribution')
plt.show()

# Pie chart for Stress Severity
stress_severity_counts = df['Stress_Severity'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(stress_severity_counts, labels=stress_severity_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Stress Severity Distribution')
plt.show()
```

Conclusion: Distribution of severity classes is not perfecly distributed, but there seem to be enough representation of each class to predict the classes.

# Finding some correlations in the dataset

Correlation between the scores

```python
# Calculate the correlation matrix
correlation_matrix = df[['Depression_Score', 'Anxiety_Score', 'Stress_Score']].corr()

# Display the correlation matrix
print(correlation_matrix)
```

```python
# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', cbar=True, square=True)
plt.title('Correlation between Depression, Anxiety, and Stress Scores')
plt.show()
```

```python
sns.pairplot(df[['Depression_Score', 'Anxiety_Score', 'Stress_Score']],
             plot_kws={'s': 5, 'alpha': 0.1},diag_kws={'bins': 43})  # 'alpha' sets the transparency level
plt.show()
```

Conclusions:
* Depression scores correlate strongly with stress score
* Anxiety score correlates storngly with stress scores
* Correlation between depression and anxiety is also existing, but weaker
