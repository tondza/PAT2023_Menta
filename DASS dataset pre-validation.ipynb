```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('/content/drive/My Drive/Data Science/DASS dataset/data.csv', delimiter='\t')

# PRINTING OUT EXAMPLE VALUES OF ALL COLUMNS

#Looking through all the columns
pd.set_option('display.max_columns', 30) # Change this to the number of columns you want to see at once

# Define the number of columns to display at a time
chunk_size = 30

# Number of chunks to iterate through
num_chunks = -(-df.shape[1] // chunk_size)  # This is a way to do ceiling division

for i in range(num_chunks):
    start_col = i * chunk_size
    end_col = start_col + chunk_size
    display(df.iloc[:, start_col:end_col].head())
```

```python
country_counts = df['country'].value_counts().head(20)
country_counts.plot(kind='bar', figsize=(10, 6))
plt.title('Frequency of Each Country')
plt.xlabel('Country')
plt.ylabel('Frequency')
plt.xticks(rotation=45)  # Rotates the country labels for better visibility if they're long
plt.show()
```

# Distribution of respondants #

```python
# Dictionary with the description of each column
description = {
    'education': {1:'Less than high school', 2:'High school', 3:'University degree', 4:'Graduate degree'},
    'urban': {1:'Rural', 2:'Suburban', 3:'Urban'},
    'gender': {1:'Male', 2:'Female', 3:'Other'},
    'engnat': {1:'Yes', 2:'No'},
    'hand': {1:'Right', 2:'Left', 3:'Both'},
    'religion': {1:'Agnostic', 2:'Atheist', 3:'Buddhist', 4:'Christian (Catholic)', 5:'Christian (Mormon)',
                6:'Christian (Protestant)', 7:'Christian (Other)', 8:'Hindu', 9:'Jewish', 10:'Muslim', 11:'Sikh', 12:'Other'},
    'orientation': {1:'Heterosexual', 2:'Bisexual', 3:'Homosexual', 4:'Asexual', 5:'Other'},
    'race': {10:'Asian', 20:'Arab', 30:'Black', 40:'Indigenous Australian', 50:'Native American', 60:'White', 70:'Other'},
    'voted': {1:'Yes', 2:'No'},
    'married': {1:'Never married', 2:'Currently married', 3:'Previously married'}
}

# For age, familysize, and major, it would be more appropriate to use histograms or similar plots.
# However, for simplicity, let's just use bar plots for the categorical columns:

for column, labels in description.items():
    plt.figure(figsize=(12,6))
    df[column].map(labels).value_counts().plot(kind='bar', color='skyblue', edgecolor='black')
    plt.title(column.capitalize())
    plt.ylabel('Count')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

# For `age` and `familysize`, histograms might be more appropriate:

for column in ['age', 'familysize']:
    plt.figure(figsize=(12,6))
    df[column].plot(kind='hist', bins=20, color='skyblue', edgecolor='black')
    plt.title(column.capitalize())
    plt.ylabel('Count')
    plt.xlabel(column)
    plt.tight_layout()
    plt.show()

# For `major`, it could be challenging to visualize due to the likely large variety of majors.
# But if you want to visualize the top 10 majors, you can use:
plt.figure(figsize=(12,6))
df['major'].value_counts().head(10).plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Major')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

# Amount of missing values in "classificator columns" #

```python
# List of columns to check
columns_to_check = ['education', 'urban', 'gender', 'engnat', 'hand', 'religion', 'orientation', 'race', 'voted', 'married', 'age', 'familysize', 'major']

# Print count and percentage of zeros for each column
print("Column\t\tCount of 0's\tPercentage of 0's")
print("-" * 50)

for column in columns_to_check:
    count_zeros = (df[column] == 0).sum()
    percentage_zeros = (count_zeros / len(df)) * 100
    print(f"{column:<15}\t{count_zeros:<12}\t{percentage_zeros:.2f}%")
```
